# Customer Journey Visualization

### Importing Data

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt

customer_journey = pd.read_csv('../data/processed/customer_journey.csv')
```

### 1.Channel Performance

```python
channel_conversion = customer_journey.groupby('Country')['CustomerID'].count().sort_values(ascending=False).head(5)

figure = go.Figure(
  go.Bar(
    x=channel_conversion.index,
    y=channel_conversion.values,
    text=(channel_conversion.values),
    textposition='auto',
    name='Channel Conversion Rate'
  )
)

figure.show()
```

### 2.Conversion Funnel

```python
funnel_periods = ['Night', 'Morning', 'Afternoon', 'Evening']
funnel_values = [
  customer_journey[customer_journey['TimeOfDay'] == period].groupby('TimeOfDay')['CustomerID'].count().values[0]
  for period in funnel_periods
]
funnel_values.sort()

figure = go.Figure(
  px.funnel_area(
    names=funnel_periods,
    values=funnel_values
  )
)

figure.show()
```
